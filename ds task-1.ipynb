```python
import matplotlib.pyplot as plt

# Sample data: ages of individuals in a population
ages = [22, 25, 27, 30, 30, 34, 35, 40, 40, 45, 47, 50, 52, 55, 57, 60, 62, 65, 70, 75]

# Define the number of bins for the histogram
num_bins = 10

# Create the histogram
plt.hist(ages, bins=num_bins, edgecolor='black')

# Add title and labels
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
